#### Importing The Libraries

```python
import cv2
import numpy as np
from PIL import Image
from IPython.display import display
```

#### Reading the Orignal Image

```python
im1 = cv2.imread("F:\Bhaiya\panda_Image.jpg")
```

```python
im1=cv2.cvtColor(im1,cv2.COLOR_BGR2RGB)
```

```python
display(Image.fromarray(im1))
```

```python
gray_image = cv2.cvtColor(im1, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(gray_image))
```

```python
inv_gray_image = 255 - gray_image
inv_gray_image
```

```python
display(Image.fromarray(inv_gray_image))
```

```python
blur_image = cv2.GaussianBlur(inv_gray_image, (35, 35), 0)
display(Image.fromarray(blur_image))
```

## Mixing Inverted Blur Image with Gray Scale Image

```python
def dodge(gray_image, blur_image):
    inv_blur_img = 255 - blur_image
    return cv2.divide(gray_image, inv_blur_img, scale = 450.0)
inv_blur_img = 255 - blur_image
```

## Enhanced Image

```python
pencil_sketch = dodge(gray_image, inv_blur_img)
display(Image.fromarray(pencil_sketch))
```
